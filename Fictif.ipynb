```python
import pandas as pd

data = {
    'nom_client': [
        'Château Margaux', 'Domaine Lefebvre', 'Cave Cooperative du Sud',
        'Vignobles Martin', 'Maison Dupont', 'Château Margaux',
        'Domaine Lefebvre', 'Les Caves Réunies', 'Vignobles Martin',
        'Maison Dupont'
    ],
    'date_commande': [
        '05/01/2024', '12/01/2024', '20/01/2021',
        '03/02/2024', '14/02/2024', '28/02/2024',
        '10/03/2024', '15/03/2024', '22/03/2024',
        '01/04/2024'
    ],
    'prix': [1200.00, 850.50, 3200.00, 450.00, 980.00,
             1500.00, 620.00, 2750.00, 390.00, 1100.00],
    'type_bouchon': [
        'Liège naturel', 'Liège aggloméré', 'Liège naturel',
        'Synthétique', 'Liège naturel', 'Liège naturel',
        'Synthétique', 'Liège aggloméré', 'Liège aggloméré',
        'Liège naturel'
    ],
    'lieu': [
        'Bordeaux', 'Bourgogne', 'Languedoc',
        'Bordeaux', 'Champagne', 'Bordeaux',
        'Bourgogne', 'Provence', 'Bordeaux',
        'Champagne'
    ]
}

client_ids = ['C001', 'C002', 'C003', 'C004', 'C005',
              'C006', 'C007', 'C008', 'C009', 'C010']

df = pd.DataFrame(data, index=client_ids)
df.index.name = 'client_id'

print(df)
df.head(10)
```

```python
df['date_commande'] = pd.to_datetime(df['date_commande'], format='%d/%m/%Y')
df['annee'] = df['date_commande'].dt.year

commandes_par_an = df.groupby(['nom_client', 'annee']).size().reset_index(name='nb_commandes')

print(commandes_par_an)
```

```python
classement = df.groupby([df.index, 'nom_client'])['prix'].sum().reset_index()
classement.columns = ['client_id', 'nom_client', 'total_depense']
classement = classement.sort_values('total_depense', ascending=False)

print(classement)
```

```python
bouchons_par_client = df.groupby('nom_client')['type_bouchon'].unique().reset_index()
bouchons_par_client.columns = ['nom_client', 'types_bouchons']

print(bouchons_par_client)
```

```python
clients_recents = df.groupby('nom_client')['date_commande'].max().reset_index()
clients_recents = clients_recents.sort_values('date_commande', ascending=False)

print(clients_recents)
```

```python
from datetime import date

SEUIL_MOIS = 12
aujourd_hui = pd.Timestamp(date.today())

derniere_commande = df.groupby('nom_client')['date_commande'].max().reset_index()
derniere_commande['jours_inactif'] = (aujourd_hui - derniere_commande['date_commande']).dt.days
derniere_commande['churn'] = derniere_commande['jours_inactif'] > (SEUIL_MOIS * 30)

print(derniere_commande.sort_values('jours_inactif', ascending=False))
```

```python
from datetime import date

SEUIL_MOIS = 12
aujourd_hui = pd.Timestamp(date.today())

derniere_commande = df.groupby('nom_client')['date_commande'].max().reset_index()
derniere_commande['jours_inactif'] = (aujourd_hui - derniere_commande['date_commande']).dt.days
derniere_commande['churn'] = derniere_commande['jours_inactif'] > (SEUIL_MOIS * 30)

taux_churn = derniere_commande['churn'].mean() * 100

print(derniere_commande)
print(f"\nTaux de churn : {taux_churn:.1f}%")
```

```python
df['date_commande'] = pd.to_datetime(df['date_commande'], format='%d/%m/%Y')

# 2. Trouver la date de première commande pour chaque client
premiere_commande = df.groupby('client_id')['date_commande'].min()

# 3. Calculer la proportion de clients depuis plus de 3 ans
date_limite = pd.Timestamp.now() - pd.DateOffset(years=3)
proportion = (premiere_commande < date_limite).mean()

print(f"Proportion de clients depuis plus de 3 ans : {proportion:.2%}")
```
